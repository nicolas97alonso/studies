
#### Access Azure Data Lake using SAS keys
1. Set the spark config SAS Token
2. List files from demo container
3. Read data from circuits.csv file 

```python
sas_key_secret = dbutils.secrets.get(scope="formula1-scope", key="formula1dl-sas-key")
```

```python
spark.conf.set("fs.azure.account.auth.type.formula1dl02.dfs.core.windows.net", "SAS")
spark.conf.set("fs.azure.sas.token.provider.type.formula1dl02.dfs.core.windows.net", "org.apache.hadoop.fs.azurebfs.sas.FixedSASTokenProvider")
spark.conf.set("fs.azure.sas.fixed.token.formula1dl02.dfs.core.windows.net", sas_key_secret)
```

```python
display(dbutils.fs.ls("abfss://demo@formula1dl02.dfs.core.windows.net"))
```

```python
df = spark.read.csv("abfss://demo@formula1dl02.dfs.core.windows.net/circuits.csv", header=True)
display(df)
```
