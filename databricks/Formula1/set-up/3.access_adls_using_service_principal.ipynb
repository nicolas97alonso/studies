
#### Access Azure Data Lake using Service Principal
1. Register Azure AD Application/ Service Principal
2. Generate a secret/password for the Application
3. Set Spatk Config with App/Client id, Directory/ Tenant Id & Secret
4. Assign Role 'Storage Blob Data Contributor' to the Data Lake 

```python
client_id = dbutils.secrets.get(scope="formula1-scope", key="formula1-app-client-id")
tenant_id = dbutils.secrets.get(scope="formula1-scope", key="formula1-app-tenant-id")
client_secret = dbutils.secrets.get(scope="formula1-scope", key="formula1-app-client-secret")
```

```python

spark.conf.set("fs.azure.account.auth.type.formula1dl02.dfs.core.windows.net", "OAuth")
spark.conf.set("fs.azure.account.oauth.provider.type.formula1dl02.dfs.core.windows.net", "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider")
spark.conf.set("fs.azure.account.oauth2.client.id.formula1dl02.dfs.core.windows.net", client_id)
spark.conf.set("fs.azure.account.oauth2.client.secret.formula1dl02.dfs.core.windows.net", client_secret)
spark.conf.set("fs.azure.account.oauth2.client.endpoint.formula1dl02.dfs.core.windows.net", f"https://login.microsoftonline.com/{tenant_id}/oauth2/token")
```

```python
display(dbutils.fs.ls("abfss://demo@formula1dl02.dfs.core.windows.net"))
```

```python
df = spark.read.csv("abfss://demo@formula1dl02.dfs.core.windows.net/circuits.csv", header=True)
display(df)
```
