```python
def define_path(container:str, storage_account)->str:
    path = f"abfss://{container}@{storage_account}.dfs.core.windows.net/"
    return path
```

```python
from pyspark.sql import DataFrame
from pyspark.sql.functions import current_timestamp
def add_ingestion_date(df:DataFrame)->DataFrame:
    return df.withColumn("ingestion_date", current_timestamp())
```
